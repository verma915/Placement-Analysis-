```python

# Count values that are either NaN or text "NA"
na_count = df['Admission Month'].isna().sum() + (df['Admission Month'].str.upper() == "NA").sum()

print("NA values in Admission Month column:", na_count)
```

```python
print("Unique Admission Month values:")
print(df['Admission Month'].unique()[:50])   # show first 50 unique values

print("\nUnique Placed Month values:")
print(df['Placed Month'].unique()[:50])      # show first 50 unique values
```

```python
# --- Admission Month Cleaning ---
def clean_admission(val):
    if pd.isna(val): 
        return pd.NaT
    if isinstance(val, pd.Timestamp):   # already a datetime
        return val
    val_str = str(val).strip().upper()
    if val_str in ["NA", "N/A", "ONGOING", ""]:
        return pd.NaT
    try:
        # Try parsing text like "AUGUST", "OCTOMBER 2024"
        return pd.to_datetime(val_str, errors='coerce')
    except:
        return pd.NaT

df['AdmissionMonthDate'] = df['Admission Month'].apply(clean_admission)

# --- Placed Month Cleaning ---
def clean_placed(val):
    if pd.isna(val): 
        return pd.NaT
    val_str = str(val).strip()
    try:
        # Handle formats like "May'25", "Jun'25", "Sept'25"
        return pd.to_datetime(val_str, format="%b'%y", errors='coerce')
    except:
        # Try full month names like "June'25"
        return pd.to_datetime(val_str, format="%B'%y", errors='coerce')

df['PlacedMonthDate'] = df['Placed Month'].apply(clean_placed)

# --- Calculate Placement Duration ---
df['PlacementDurationMonths'] = (df['PlacedMonthDate'] - df['AdmissionMonthDate']).dt.days / 30

# --- Flag Placements within 6 months ---
df['PlacedWithin6Months'] = df['PlacementDurationMonths'] <= 6

# --- Summary ---
total_students = len(df)
placed_students = df['PlacedMonthDate'].notna().sum()
ongoing_students = df['PlacedMonthDate'].isna().sum()
placed_within_6 = df['PlacedWithin6Months'].sum()
percent_within_6 = (placed_within_6 / placed_students * 100) if placed_students > 0 else 0

print("Total Students:", total_students)
print("Placed Students:", placed_students)
print("Ongoing Students:", ongoing_students)
print("Placed within 6 months:", placed_within_6)
print("Percentage placed within 6 months:", round(percent_within_6, 2), "%")
```
